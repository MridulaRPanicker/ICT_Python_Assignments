#**Numpy Problems**

# 1. Basic Array Operation
 Convert the mpg column into a NumPy array and calculate:

○ The mean, median, and standard deviation of mpg.

○ The number of cars with mpg greater than 25.

```python
import numpy as np # importing numpy as np to use.
```

```python
# loading the entire auto-mpg csv file data into a Structured Array using np.genfromtxt

auto_mpg = np.genfromtxt('/content/drive/MyDrive/ICT_DSA_Course_Programs/Data_toWork/auto-mpg.csv',\
                         delimiter=',',dtype=None, names=True,encoding='utf-8')
#print(auto_mpg.ndim)
```

```python
# The mean, median, and standard deviation of mpg.

mean_mpg = np.round(np.mean(auto_mpg['mpg']),2) # Gives the mean mpg value
median_mpg = np.round(np.median(auto_mpg['mpg']),2) # Gives the median mpg value
stdev_mpg = np.round(np.std(auto_mpg['mpg']),2) # Gives the std deviation mpg value

print(f"The Mean mpg of cars = {mean_mpg}\
\nThe Meadian mpg of cars = {median_mpg}\
\nThe Standard Deviation mpg of cars = {stdev_mpg}")
```

```python
# The number of cars with mpg greater than 25.

count_mpg = np.count_nonzero(auto_mpg['mpg']>25)
print(f"Number of cars with mpg more than 25 = {count_mpg}")
```

#2. Filtering
Using NumPy, filter all cars with more than 6 cylinders.

Return the corresponding car_name as a list.

```python
# filtering of cars based on muber of cylinders
# picking the car name based on the conditions satisfied

sixPlus_cylinder = auto_mpg[auto_mpg['cylinders']>6]['car_name']

# while printing converted the ndArray to a list
# printed only the 1st 5 filtered car names in the form of a column

print(*sixPlus_cylinder.tolist()[:5],sep='\n')
```

#3. Statistical Analysis

Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy.

```python
# Calculated the percentile and saved it in variables
p25,p50,p75 = np.percentile(auto_mpg['weight'],[25,50,75])

print(f"25th Percentile of Weight = {p25}\
        \n50th Percentile of Weight = {p50}\
        \n75th Percentile of Weight = {p75}")
```

#4. Array Manipulation
Convert the acceleration column into a NumPy array and normalize its values (scale between 0 and 1).

```python
auto_acceleration = auto_mpg['acceleration'] # Got variable Acceleration to an array

# Normalised each datapoint in the accelation between 0 & 1 using the formula
# (datapoint-min(acceleration))/(max(acceleration)-min(acceleration))

normalised_accel = np.round((auto_acceleration-np.min(auto_acceleration))/np.ptp(auto_acceleration),2)

# Printed 1st 5 normalised datapoints as a sample in a column format
print(*normalised_accel[:5],sep = '\n')
```

#5. Broadcasting
Increase all horsepower values by 10% and store the updated values in a new NumPy array. Handle missing data (if any) by replacing it with the mean of the column before applying the increase.

```python
horsepower_clean = auto_mpg['horsepower'] # Got the horsepower to an array

# identified '?' as missing values and replaced it with
# python understandable string 'nan'
horsepower_clean[horsepower_clean == '?'] = 'nan'

# Converted all the values in the array to float datatype
auto_horsepower = np.array(horsepower_clean, dtype=float)

# Calculated mean horse power value
mean_horsepower = np.nanmean(auto_horsepower)

# Replaced all 'nan' with mean value
auto_horsepower[np.isnan(auto_horsepower)] = mean_horsepower
print(auto_horsepower[:5]) #printed sample 5 values from original horsepower array

# Increased horsepower values by 10%
horsepower_increased = auto_horsepower * 1.1
print(horsepower_increased[:5]) #printed sample 5 values from increased horsepower array
```

#7. Matrix Operations
Create a 2D NumPy array containing the columns mpg, horsepower, and weight. Compute the dot product of this matrix with a given vector [1, 0.5, -0.2].

```python
# Read the csv file as a 2D array

auto_mpg_2D = np.genfromtxt('/content/drive/MyDrive/ICT_DSA_Course_Programs/Data_toWork/auto-mpg.csv',\
                            delimiter=',',dtype=float,skip_header=1,filling_values='nan' ,encoding='utf-8')

# Created a new 2D array with mpg,horsepower and weight
# Got a 398 x 3 matrix
mpg_hp_weight = auto_mpg_2D[:,[0,3,4]]

# Stored given vector in a variable as a list
vec_given = [1,0.5,-0.2]

# Performed dot product using matrix and vector
# Resulted in a 1D array with 398 rows
dotproduct = np.round(mpg_hp_weight @ vec_given,2)

# Printed the first 5 rows as a sample in a column form
print(*dotproduct[:5],sep='\n')
```

#10. Conditional Aggregates
Calculate the mean mpg for cars grouped by the number of cylinders using NumPy techniques.

```python
# Identify unique cylinders and their mapping with all the other values
#return_inverse gives an array of indices to build the original array later
unique_cylinders, inverse = np.unique(auto_mpg['cylinders'],return_inverse=True)

# converted the unique cylinders to a list for final o/p printing
unique_cyl_list = unique_cylinders.tolist()

# used np.bincount along with weights helps to get the sum of mpg based on each cylinder bins
mpg_sums = np.bincount(inverse, weights=auto_mpg['mpg'])
#counted the number of values in each bin
mpg_count = np.bincount(inverse)

# calculated mean using the sum & count then converted it into a list
mean_mpg_cyl = np.round(mpg_sums/mpg_count,2)
mean_tolist = mean_mpg_cyl.tolist()

final = dict(zip(unique_cyl_list,mean_tolist))
print(final)
```

#10. Finding Anomalies
Identify potential outliers in the mpg column using the Interquartile Range (IQR) method. Specifically:

● Calculate the IQR for mpg.

● Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * IQR.

● Create a DataFrame of cars classified as outliers, displaying car_name, mpg, and model_year.


```python
# Calculated Q1 & Q3 using pandas quantile method
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)

# calculated IQR and printed the result
IQR = Q3-Q1
print(f"IQR of mpg = {IQR}")
```

```python
# identified lower and upper threashold points
# beyond this the datapoints will be trated as outliers
lower_outlier = Q1 -1.5 * IQR
upper_outlier = Q3 + 1.5 * IQR

print(f"Lower Threashold value = {lower_outlier}\
\nUpper Threashold value = {upper_outlier}")
```

```python
# identified outlier rows based on lower and upper threashold
outlier_df = df[(df['mpg']<lower_outlier) | (df['mpg']>upper_outlier)]

#printed specif columns from the outlier DataFrame
print(outlier_df[['car name','mpg','model year']])
```

2.1 Calculate the total revenue for the store (sum of all quantities × price per item).

```python
Total_revenue = 0
for item in sales_data_ShopEase:
  Total_revenue += item['quantity'] * item['price_per_item']

print(f"Total Revenue for ShopEase is {Total_revenue}")
```

2.2 Identify all orders where the total order value (quantity × price per item) exceeds ₹2,000. Store the result as a new list.

```python
sales_above_twoThousand = []
for item in sales_data_ShopEase:
  Total_order_val = item['quantity'] * item['price_per_item']
  if Total_order_val > 2000:
    sales_above_twoThousand.append(list(item.values()))
  else:
    continue

for item in sales_above_twoThousand:
  print(item)
```

2.3 Find the item with the highest total quantity sold across all orders.

```python
total = {}
for i in sales_data_ShopEase:
  item = i['item']
  total[item] = total.get(item, 0) + i['quantity']

high_item = max(total, key = total.get)
print(f"Item with highest quantity sold: {high_item} ({total[high_item]} units)")

```

Write a function apply_discount(orders, discount_rate) that takes a list of orders and applies a percentage discount (e.g., 10% = 0.10) to the price_per_item for each order.

```python
def apply_discount(orders, discount_rate):
  new_discount_list = []
  for order in orders:

        new_order = dict(order)

        original_price = new_order['price_per_item']
        new_order['price_per_item'] = original_price * (1 - discount_rate)

        new_discount_list.append(new_order)

  return new_discount_list

discounted_sales = apply_discount(sales_data_ShopEase, 0.10)

for order in discounted_sales:
  print(list(order.values()))

```

4.1 Write a custom iterator class OrderIterator to iterate through the orders, yielding one order at a time.

```python
class OrderIterator():
  def __init__(self, orders):
    self._orders = orders
    self._index = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self._index < len(self._orders):
      result = self._orders[self._index]
      self._index +=1
      return result
    else:
      raise StopIteration

order_iterat = OrderIterator(sales_data_ShopEase)

print("Iterating through orders:")
for val in order_iterat:
  print(f"Order ID: {val['order_id']} | Item: {val['item']}")
```

Create a dictionary grouping all orders by customer using a dictionary comprehension.

```python
grouped_orders = {
    customer: [order for order in sales_data_ShopEase if order['customer'] == customer]
    for customer in {order['customer'] for order in sales_data_ShopEase}
}


for customer, orders in grouped_orders.items():
    print(f"{customer}: {len(orders)} order(s)")
```

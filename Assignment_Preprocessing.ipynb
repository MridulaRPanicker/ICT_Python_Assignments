####**Handling Missing Values**

```python
# Identifyting missing data from each features and providing the result in total and percentage as well. The final result in sorted format.

missing_data = (pd.concat([data.isna().sum(),(data.isna().sum()/
                                              len(data))*100],axis = 1,keys=['Total Missing Vals','% Missing']).sort_values
                                               (by='% Missing',ascending=False))
print(missing_data)
```

1. Handling feature 'No of Times Visited'.
  - This feature has the highest number of missing values.
  - The above report shows 90% of the data missing. Trying to impute this high missing value rate is going to affect the entire dataset. So decided to drop the feature.

```python
data.columns
```

```python
data.drop(['No of Times Visited'],axis = 1, inplace = True) # removed the feature 'No of times visited'
data.shape # shows the new dataframe with one feature less
```

2. Handling feature 'Sale Price'
  - This is the dependent feature. This feature has 4 missing values.
  - But altering the data in the dependent feature can impact the algorithm by a huge margin and it can create bias in the prediction.
  - Since the missing values are a very small %, decided to drop those records to make the dataset clean for analysis.

```python
#missing_sale_price = data[data['Sale Price'].isna()]
#print(missing_sale_price)

data.dropna(subset=['Sale Price'],inplace = True) # Dropped 4 records with missing value in target feature

print(f"Missing Values in Sale Price: {data['Sale Price'].isna().sum()}") # Ensure dataset doesn't have any missing value in target feature
data.shape  # Shape of the dataset now changed
```

3. Handling missing value from the numerical features.
 -  Flat Area (in Sqft),Lot Area (in Sqft), No of Bathrooms,Area of the House from Basement (in Sqft),Living Area after Renovation (in Sqft)
 - Above mentioned are the numerical features and in these features, it is best to impute with 'median' rather than 'mean' to avoid extreme value influence.


```python
numerical_features = ['No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)',
       'Area of the House from Basement (in Sqft)','Living Area after Renovation (in Sqft)']
for feature in numerical_features:
  median_val = data[feature].median()
  data[feature].fillna(median_val, inplace=True)

print(data[numerical_features].isna().sum())
```

4. Handling missing value in non-numerical features.
 -  Latitude, Longitude and Zip Code are presented as numerical features. But it is better to impute with 'mode' to keep it consistent with data provided

```python
categoric_features = ['Zipcode', 'Latitude','Longitude']
for feature in categoric_features:
  mode_val = data[feature].mode()[0]
  data[feature].fillna(mode_val, inplace=True)

print(data[categoric_features].isna().sum())
```

####**Scaling Numerical Variables**

- For scaling, initially need to isolate all the numerical features.
- Choosing Standard scaling method to perform the scaling.
- This scaling help to avoid bias by the algorithm due to scale of different features.
- Scaling applied to only logically standing numerical features. Hence avoided ID,Dates,Zipcode,latitude and longitude.


```python
from sklearn.preprocessing import StandardScaler

num_features = ['No of Bedrooms','No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)',
       'No of Floors', 'Overall Grade', 'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
       'Age of House (in Years)','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)']

scaler = StandardScaler()

data[num_features] = scaler.fit_transform(data[num_features])

data.head()
```

####**Encoding Categorical Variables**

Feature 'Condition of the House' is with values ranging from Bad to Fair. So we can use Label Encoder to handle this feature.

```python
data['Condition of the House'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder # Encoding process

le = LabelEncoder()
data['Condition of the House'] = le.fit_transform(data['Condition of the House'])
```

Feature 'Waterfront View' will get transformed use OHE

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first')

# Fit and transform the waterfront column
encoded_cols = encoder.fit_transform(data[['Waterfront View']])

# Convert back to a DataFrame and join
encoded_data = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(['Waterfront View']))
data = data.join(encoded_data).drop('Waterfront View', axis=1)
```

```python
data['Waterfront View_Yes'].value_counts()
```

Need to transform 'Date House was sold' to a meaningful datetime feature and from that need to extract date and year.

```python
data['Date House was Sold'] = pd.to_datetime(data['Date House was Sold'])
data['Selling Year'] = data['Date House was Sold'].dt.year
data['Selling Month'] = data['Date House was Sold'].dt.month
data.drop(columns=['Date House was Sold'], inplace=True)
```

'Renovated Year' is also need to be modified to make it logically inplace with the algorithm. So going to convert the feature to 'Renovated' or 'Not renovated' and this will consider when the reniovation happened as well.

```python
data['Is Renovated'] = data['Renovated Year'].apply(lambda x: 1 if x > 0 else 0)

# Years since last update (uses Yr_Built if never renovated)
data['Years Since Update'] = data['Selling Year'] - data[['Renovated Year', 'Age of House (in Years)']].max(axis=1)
data['Years Since Update'] = scaler.fit_transform(data[['Years Since Update']]) # scaled to fix the issue
```

####**Outlier Removal**




Checking outliers in feature Sale Price since that affects the prediction logic

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=data['Sale Price'])
plt.title('Boxplot of Numerical Features')
plt.show()
```

```python
Q1 = data['Sale Price'].quantile(0.25)
Q3 = data['Sale Price'].quantile(0.75)
IQR = Q3 - Q1

# Define boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataset
data = data[(data['Sale Price'] >= lower_bound) & (data['Sale Price'] <= upper_bound)]
```

# 1. Bank Account Manager

```python
account_balance = 1000 # set an initial amount of 1000 to the account


# function to deposit to the existing account
def account_deposit(val):
    global account_balance
    if val>0:               # deposit is allowed only if the depositing amount is more than 0
      account_balance = account_balance + val
      print(f"Amount {val} deposited. Available balance is {account_balance}")
    else:
      print("Kindly enter an amount more than zero for deposit")


# function to withdraw amount
def account_withdraw(val):
    global account_balance
    if val > account_balance:
      print("Transaction failed. Insufficiant balance.")
    elif val<=0:
      print("Kindly enter an amount more than zero for withdrawal")
    else:
      account_balance = account_balance-val
      print(f"{val} debited from account current balance is {account_balance}")


# function to check the balance
def balance_enquiry():
    global account_balance
    print(f"Available balance in your account is {account_balance}")

account_deposit(100)
account_withdraw(350)
balance_enquiry()
account_deposit(500)
balance_enquiry()
account_withdraw(1500)
```

# 2. Weather Data Analyzer

```python
# Provided are the temperature, humidity and cloud cover of TVM in the last 5days

region = "Thiruvananthapuram"
temperature = [28,29,25,9,22]
humidity_percent = [50,55,60,88,90]
cloudCover_percent = [12,20,35,80,90]



# Calculate average weather of TVM for the given days
def average_weather(input_data):
    if not input_data:
      print("No weather data provided")
    else:
      avg_weather = round(sum(input_data)/len(input_data),2)
      return avg_weather


# Define weather condition based on the given data
def weather_condition(temp,hum,cloud_cov):
    if hum<60 and 0<cloud_cov<12.5:
        return "Clear/Sunny"
    elif temp>27 and hum<60 and 0<cloud_cov<25:
        return "Hot climate and clear to sunny"
    elif 18<temp<27 and 50<hum<70 and 25<cloud_cov<60:
        return "Pleasant climate with partly cloudy/sunny"
    elif temp<18:
        return "Cold climate"
    elif hum>80 and 60<cloud_cov<100:
        return "Rainy climate with cloudy/overcast sky"


# Print the daily weather condition of region based on given data
print(f"Daily weather condition in {region} for the past {len(temperature)} days")
for i in range(len(temperature)):
    weather = weather_condition(temperature[i],humidity_percent[i],cloudCover_percent[i])
    print(f"Day {i+1}: Temperature {temperature[i]}°C, Humidity {humidity_percent[i]}%, Cloud Cover {cloudCover_percent[i]}% and it is {weather}")


# Average weather condition for the given number of days
average_temp = average_weather(temperature)
print(f"The Average Temperature in {region} for the past {len(temperature)} days is {average_temp}°C.")
average_hum = average_weather(humidity_percent)
print(f"The Average Humidity in {region} for the past {len(humidity_percent)} days is {average_hum}%.")


# Check weather condition for an input data
print(weather_condition(9,80,80))
print(weather_condition(25,55,50))


```
